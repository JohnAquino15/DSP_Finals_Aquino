**Part 1.** [200 pts] Compute for the Energy and Power of the following signals in terms of per unit use the
proper formula whether the signal is continuous or discrete.


```python
import numpy as np
import scipy as sp
import sympy as smp
import matplotlib.pyplot as plt
```

**a.** $ \sin[n]$

```python
x = smp.symbols('n', real=True)
f = smp.sin(x)
E= smp.integrate(f, x)
print("E =")
E
```

```python
T=np.pi
t=np.linspace(-T, T, 1000)
E= smp.integrate(f, x)
P = (1/(2*T))*(smp.limit(E,x,T) + smp.limit(E,x,-T))
print("P =")
P
```

```python
E_values = np.array([E.subs(x, val) for val in t], dtype='complex')

# Calculate the power using the limits
P = (1 / (2 * T)) * (smp.limit(E, x, T) + smp.limit(E, x, -T))

# Plot the real and imaginary parts of f
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, np.real(E_values), label='Real part')
plt.plot(t, np.imag(E_values), label='Imaginary part')
plt.title(f'Real and Imaginary Parts of f\n$E = {smp.latex(E)}$')
plt.legend()

# Plot the power P
plt.subplot(2, 1, 2)
plt.axhline(y=P.evalf().as_real_imag()[0], color='r', linestyle='--', label=f'Signal Power\n$P = {smp.latex(P)}$')
plt.title(f'Signal Power\n$P = {smp.latex(P)}$')
plt.legend()

plt.tight_layout()
plt.show()
```

**Part 2.** [100 pts] Prove the equality of the following trigonometric functions using Euler identities. Further
prove the equality in Python. Furthermore, plot the left-hand side and right-hand side of the equation and
their Euler forms as additional proof.


$ (sin^2(\theta) = \frac{1}{2} (1-cos(2\theta)))$

to simplifiy, multiply both sides by 2

$ 2sin^2(\theta) = \frac{1}{2} (1-cos(2\theta))2$

simplified: $ 2sin^2(\theta) = 1-cos(2\theta)$

euler's indentity: $ e^{i\pi} + 1 = 0 $


```python
x = smp.symbols('n', real=True)
a = 2*(smp.sin(x)*smp.sin(x))
b = 1 -  smp.cos(2*x)
```

```python
a
```

```python
b
```

```python
# Simplify both expressions
a_simplified = smp.simplify(a)
b_simplified = smp.simplify(b)

# Check if the two simplified expressions are equal
if a_simplified == b_simplified:
    print("The equality is proven.")
else:
    print("The equality is not proven.")
```

```python
#If the difference is zero it is equal
a_simplified - b_simplified
```

```python
T =np.pi
# Create numerical functions for plotting
f_a = smp.lambdify(x, a, 'numpy')
f_b = smp.lambdify(x, b, 'numpy')

# Generate x values
x_values = np.linspace(-T, T, 1000) 
# Plot the expressions
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.plot(x_values, f_a(x_values), label='sin(x)^2', color = 'red')
plt.title('Left-Hand Side')
plt.xlabel('x')
plt.ylabel('Expression Values')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_values, f_b(x_values), label='1 - cos(2x)')

plt.title('Right-Hand Side')
plt.xlabel('x')
plt.ylabel('Expression Values')
plt.legend()
plt.grid(True)
plt.show()
```

**To prove it they are EQUAL, when plotted together the lines should OVERLAP**

```python
plt.plot(x_values, f_a(x_values), label='sin(x)^2', color = 'red')
plt.plot(x_values, f_b(x_values), label='1 - cos(2x)')
plt.title('Comparison')
plt.xlabel('x')
plt.ylabel('Expression Values')
plt.legend()
plt.grid(True)
```

$ sin^2(\theta) = \frac{1}{2} (1-cos(2\theta))$ , **EQUALITY PROVEN**


**Part 3.** [100 pts] Prove the equality of the following trigonometric functions using Euler identities. Further
prove the equality in Python (NumPy). 


$ e^{i\pi}-1 =0$


```python
# Define pi and the imaginary unit i
pi = np.pi
i = 1j  # 1j represents the imaginary unit

# Euler's identity: e^(i*pi) + 1 = 0
result = np.exp(i * pi) + 1

# Check if the result is close to 0 (considering floating-point precision)
is_identity_verified = np.isclose(result, 0)

# Print the result and whether Euler's identity is verified
print(f"e^(i*pi) + 1 = {result}")
print(f"Euler's identity verified: {is_identity_verified}")
```
